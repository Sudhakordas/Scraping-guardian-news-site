# 1. Finding Category from the URL

```python
News_Category = {
                 'News-Business' : ('business'),
                 'News-Technology' : ('technology'),
                 'News-Global-Development': ('global-development'),
                 'News-Science':('science'),
                 'News-Environment':('environment'),
                 'News-international': ('world', 'australia-news', 'uk-news', 'us-news'),
                 'News-science': 'Science',
                 'Politics' : 'politics',
                 'Culture':('books', 'music', 'tv-and-radio', 'artanddesign', 'film', 'stage', 'games'),
                 'Sports':('football', 'sport'),
                 'Opinion':('commentisfree'),
                 'Lifestyle':('food', 'lifeandstyle', 'travel', 'fashion','money')}
```

```python
def spliting(url):
    return url.split('/')[3]
    
```

```python
df['base-word'] = df['Url'].apply(spliting)
```

```python
df.tail(50)
```

```python
base_word = df['base-word'].values.tolist()
```

```python
Category = []
for base in base_word:
    for key, values in News_Category.items():
        if base in values:
            Category.append(key)
            break
    else:
        Category.append('Others')
    
```

```python
df['Category'] = Category
```

```python
df['Category'].value_counts()
```

```python
Category = df['Category'].unique()
Categoty = Category.tolist()
```

# 3. Most present word in the each Category

```python
News = []
```

```python
for i, cat in enumerate(Category):
    news = ''
    ran = (df['Content'].where(df['Category'] == cat)).to_list()
    for item in ran :
        if not(pd.isnull(item)) == True:
            news = news + item
    News.append((cat,news))
    print(cat, 'Is done')
    print("\n")
```

```python
News
```

```python
Most_frequent = []
for x, y in News:
    print(x)
    word_list =  y.split(' ')
    word_counter = {}
    for word in word_list:
        if word in word_counter:
            word_counter[word] += 1
        else:
            word_counter[word] = 1
    popular_words = sorted(word_counter, key = word_counter.get, reverse = True)
    Most_frequent.append((x, popular_words[:200]))

    
```

# 4. Text Clustering Using k-means

```python
import re
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
#Text Preprocessing
lemma = WordNetLemmatizer()
def textpreprocess(text):
    Comment = re.sub('[^a-zA-Z]', ' ', str(text))
    Comment = Comment.lower()
    Comment = Comment.split()
    Comment = [lemma.lemmatize(word) for word in Comment if word not in stop_words]
    Comment = ' '.join(Comment)
    
    return Comment
```

```python
df['Clean_text'] = df['Content'].apply(textpreprocess)
```

```python
Vector = TfidfVectorizer()
X = Vector.fit_transform(df['Clean_text'])
```

```python
from sklearn.cluster import KMeans
cs = []
for i in range(1, 20):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 20), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Error')
plt.show()
```

Here we are seeing that, this is not acttually looks like albow, Hence lets Consider K = 12

```python
true_k = 12
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)

print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = Vector.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :20]:
        print(' %s' % terms[ind])
    print("\n")

 
```
